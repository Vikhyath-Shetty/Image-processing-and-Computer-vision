### Perspective Projection

```python
import cv2 as cv
import matplotlib.pyplot as plt
import os

def imshow(title,image,size=10):
    w,h = image.shape[1],image.shape[0]
    aspect_ratio = w/h
    plt.figure(figsize=(size*aspect_ratio,size))
    plt.title(title)
    plt.imshow(cv.cvtColor(image,cv.COLOR_BGR2RGB))
    plt.show()
```

```python
img = cv.imread(os.path.join("..","data","page.jpeg"))
imshow("Image",img,5)
```

```python
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
imshow('Grayscale',gray,5)
```

```python
gray = cv.GaussianBlur(gray,(5,5),0)
imshow('Gaussian Blur',gray,5)
```

```python
#Thresholding
thresh_img = cv.adaptiveThreshold(gray,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY,3,0)
imshow("Thresholded image",thresh_img,5)
```

```python
edged = cv.Canny(thresh_img,100,200)
imshow('Edged',edged,5)
```

```python
contours,heirarchy = cv.findContours(edged,cv.RETR_EXTERNAL,cv.CHAIN_APPROX_SIMPLE)
img_cpy = img.copy()
cv.drawContours(img_cpy,contours,-1,(0,255,0),2)
imshow('Contours',img_cpy,5)
print(len(contours))
```
